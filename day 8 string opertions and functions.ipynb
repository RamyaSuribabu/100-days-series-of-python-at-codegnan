## substring index finding

```python
# find()
s=("my name is ramya")
print(s.find("i"))
print(s.find("i",6)) #  6 means it starts to check the index from index 6
print(s.rfind("a"))
 # print(s.index("z")) # it gives  value error
print(s.find("z")) # it gives the -1 as index
# strip() - it removes the white space
s=("      codeganan            ")
print(s)
print(s.strip())
print(s.lstrip()) # it removes the left side space
print(s.rstrip()) # it removes the right side space
j=("@@@@code@@ganan@@@")
print(j.strip('@')) # it removes thst symbol at the ends of the string
```

```python
# split(sep)- it returns the list of substring and it separates the string with sep
j=("@@@@code@@ganan@@@")
print(j)
print(j.split('@'))
print(j.split())
print(j.rsplit())
```

```python
# case checking functions it returns true or false
s1="CODE"
s2="hnan"
s3="123"
s4="abc123"
s5="codefehbj"
print(s1.isupper())
print(s2.isupper())
print(s1.islower())
print(s2.islower())
print(s3.isdigit()) # if the string contains the space that time it gives false
print(s4.isdigit())
print(s4.isalnum())  # if the string contains the space that time it gives false
print(s4.isalpha())
print(s5.isalpha())
```
