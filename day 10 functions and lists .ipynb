```python

#checking whether the number is even or odd

def checking(a):
    if a%2 == 0:
        return True
    return False
n=int(input())
if checking(n):
    print(n,"is an even  number ")
else:
    print(n,"is a odd number ")
```

# subtraction of  2 numbers

```python

#Using positional arguments

def sub(a,b):
    res=a-b
    return res
n1=int(input())
n2=int(input())
print(sub(n1,n2)) # if we change like (n1,n2) then in def function it assigns like n1=a and n2=b
print(sub(n2,n1))# if we change like (n2,n1) then in def function it assigns like n2=a and n1=b

```

```python
def sub(a,b):
    res=a-b
    return res
n1,n2=10,5         #we can define the multiple inputs in a simgle line
print(sub(n1,n2)) # if we change like (n1,n2) then in def function it assigns like n1=a and n2=b
print(sub(n2,n1))
```

```python
#Using keyword arguments
def sub(a,b):
    res=a-b
    return res
n1,n2=10,3
print(sub(a=n1,b=n2)) # we are passing the arguments using parameter names
print(sub(n2=b,n1=a))
```

#list operations


```python
# accessing the list elements using index position
list1=[1,2.3,8,'hi',0.88,"hello"]
print(list1[0])
print(list1[-1])
print(list1[2:])
print(list1[-1:])
print(list1[:-1])
```

#read list of numbers and if number is even add 0.5 and if number is odd add 1

```python
##lst=[1,2,3,4,5]
lst=list(map(int,input().split()))
for i in range(0,len(lst)):
    if lst[i]%2 :
        lst[i]=lst[i]+1
    else:
         lst[i]=lst[i]+0.5
print(lst)
```
