```python
def add_three_numbers(a, b, c):
    return a + b + c

# Taking input from the user
num1 = int(input("Enter first number: "))
num2 = int(input("Enter second number: "))
num3 = int(input("Enter third number: "))

# Function call
result = add_three_numbers(num1, num2, num3)

print("Sum of three numbers is:", result)
```

```python
# default arguments  
def add(a,b):
    return a+b
print(add(10,2))
```

```python
#default arguments
def add(a,b,c=0,d=0):
    return a+b+c+d
print(add(10,2))
print(add(1,2,3))
print(add(1,2,3,4))
```

```python
# multiplication of numbers using default arguments
def mul(a,b,c=1,d=1):
    return a*b*c*d
print(mul(10,2))
print(mul(1,2,3))
print(mul(1,2,3,4))
```

```python
# variable length of positional arguments
def add(a,*args):
    print(a)
    print(args)
add(1,2,3,4,5,6)
```

```python
# variable length of positional arguments
a,*b,c=10,20,30,40,50
print(a)
print(b)
print(c)

```

```python
a,*b=10,20,30,40,50
print(a)
print(b)
```

```python
# variable length of positional arguments
def add(a,*args):
    print(a)
    print(args)
    s=0
    for num in args:
        s+=num
    return s
print(add(1,2,3,9))
```
